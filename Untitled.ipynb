```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
def clustering(n_clusters,df):
    for i in n_clusters:
        clusters = KMeans(n_clusters=i,random_state=42)
        clusters.fit_predict(df)
        inertias.append(clusters.inertia_)
        labels.append(clusters.labels_)
        s_scores.append(silhouette_score(df,clusters.labels_))
    return n_clusters

def plot_elbow(n_clusters):
    plt.plot([i for i in n_clusters],inertias)
    plt.xlabel("Number of Clusters")
    plt.ylabel("Inertias")
    plt.show()

def plot_silhouette(n_clusters):
    plt.plot([i for i in n_clusters],s_scores)
    plt.xlabel("Number of Clusters")
    plt.ylabel("Silhouette Score")
    plt.show()
```
