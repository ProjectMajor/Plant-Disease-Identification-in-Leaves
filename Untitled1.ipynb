```python
import pandas as pd
import numpy as np
```

```python
import cv2 as cv
from skimage.feature import graycomatrix,graycoprops
```

```python
def get_glcm_csv(path_df):
    
    """
    Takes in image directory paths as df then applies one distance to one angle not a list
    """
    contrast_1_0, energy_1_0, homogeneity_1_0, correlation_1_0 = [], [], [], []
    contrast_1_90, energy_1_90, homogeneity_1_90, correlation_1_90 = [], [], [], []
    contrast_3_0, energy_3_0, homogeneity_3_0, correlation_3_0 = [], [], [], []
    contrast_3_90, energy_3_90, homogeneity_3_90, correlation_3_90 = [], [], [], []
    contrast_5_0, energy_5_0, homogeneity_5_0, correlation_5_0 = [], [], [], []
    contrast_5_90, energy_5_90, homogeneity_5_90, correlation_5_90 = [], [], [], []
    contrast_1_45, energy_1_45, homogeneity_1_45, correlation_1_45 = [], [], [], []
    contrast_3_45, energy_3_45, homogeneity_3_45, correlation_3_45 = [], [], [], []
    contrast_5_45, energy_5_45, homogeneity_5_45, correlation_5_45 = [], [], [], []
    label = []
    data = pd.DataFrame()
    
    for i in path_df:
#         print(i[0],i[1])
#         img_path = os.path.join(i[0])
        image = cv.imread(i[0])
#         print(image)
        label.append(i[1])
    
        gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
        
        # distance 1 and angle 0
        
        glcm = graycomatrix(gray, [1], angles=[0], levels=256, symmetric=True, normed=True)
        
        contrast = graycoprops(glcm, prop="contrast")
        energy = graycoprops(glcm, prop='energy')
        homogeneity = graycoprops(glcm, prop='homogeneity')
        correlation = graycoprops(glcm, prop='correlation')
        
        contrast_1_0.append(contrast[0][0])
        energy_1_0.append(energy[0][0])
        homogeneity_1_0.append(homogeneity[0][0])
        correlation_1_0.append(correlation[0][0])
        
        # distance 1 and angle 90
        
        glcm = graycomatrix(gray, [1], angles=[90], levels=256, symmetric=True, normed=True)
        
        contrast = graycoprops(glcm, prop="contrast")
        energy = graycoprops(glcm, prop='energy')
        homogeneity = graycoprops(glcm, prop='homogeneity')
        correlation = graycoprops(glcm, prop='correlation')
        
        contrast_1_90.append(contrast[0][0])
        energy_1_90.append(energy[0][0])
        homogeneity_1_90.append(homogeneity[0][0])
        correlation_1_90.append(correlation[0][0])
        
        # distance 1 and angle 45
        
        glcm = graycomatrix(gray, [1], angles=[45], levels=256, symmetric=True, normed=True)
        
        contrast = graycoprops(glcm, prop="contrast")
        energy = graycoprops(glcm, prop='energy')
        homogeneity = graycoprops(glcm, prop='homogeneity')
        correlation = graycoprops(glcm, prop='correlation')
        
        contrast_1_45.append(contrast[0][0])
        energy_1_45.append(energy[0][0])
        homogeneity_1_45.append(homogeneity[0][0])
        correlation_1_45.append(correlation[0][0])
        
        # distance 3 and angle 0
        
        glcm = graycomatrix(gray, [3], angles=[0], levels=256, symmetric=True, normed=True)
        
        contrast = graycoprops(glcm, prop="contrast")
        energy = graycoprops(glcm, prop='energy')
        homogeneity = graycoprops(glcm, prop='homogeneity')
        correlation = graycoprops(glcm, prop='correlation')
        
        contrast_3_0.append(contrast[0][0])
        energy_3_0.append(energy[0][0])
        homogeneity_3_0.append(homogeneity[0][0])
        correlation_3_0.append(correlation[0][0])
        
        # distance 3 and angle 90
        
        glcm = graycomatrix(gray, [3], angles=[90], levels=256, symmetric=True, normed=True)
        
        contrast = graycoprops(glcm, prop="contrast")
        energy = graycoprops(glcm, prop='energy')
        homogeneity = graycoprops(glcm, prop='homogeneity')
        correlation = graycoprops(glcm, prop='correlation')
        
        contrast_3_90.append(contrast[0][0])
        energy_3_90.append(energy[0][0])
        homogeneity_3_90.append(homogeneity[0][0])
        correlation_3_90.append(correlation[0][0])
        
        # distance 3 and angle 45
        
        glcm = graycomatrix(gray, [3], angles=[45], levels=256, symmetric=True, normed=True)
        
        contrast = graycoprops(glcm, prop="contrast")
        energy = graycoprops(glcm, prop='energy')
        homogeneity = graycoprops(glcm, prop='homogeneity')
        correlation = graycoprops(glcm, prop='correlation')
        
        contrast_3_45.append(contrast[0][0])
        energy_3_45.append(energy[0][0])
        homogeneity_3_45.append(homogeneity[0][0])
        correlation_3_45.append(correlation[0][0])
        
        # distance 5 and angle 0
        
        glcm = graycomatrix(gray, [5], angles=[0], levels=256, symmetric=True, normed=True)
        
        contrast = graycoprops(glcm, prop="contrast")
        energy = graycoprops(glcm, prop='energy')
        homogeneity = graycoprops(glcm, prop='homogeneity')
        correlation = graycoprops(glcm, prop='correlation')
        
        contrast_5_0.append(contrast[0][0])
        energy_5_0.append(energy[0][0])
        homogeneity_5_0.append(homogeneity[0][0])
        correlation_5_0.append(correlation[0][0])
        
        # distance 5 and angle 90
        
        glcm = graycomatrix(gray, [5], angles=[90], levels=256, symmetric=True, normed=True)
        
        contrast = graycoprops(glcm, prop="contrast")
        energy = graycoprops(glcm, prop='energy')
        homogeneity = graycoprops(glcm, prop='homogeneity')
        correlation = graycoprops(glcm, prop='correlation')
        
        contrast_5_90.append(contrast[0][0])
        energy_5_90.append(energy[0][0])
        homogeneity_5_90.append(homogeneity[0][0])
        correlation_5_90.append(correlation[0][0])
        
        # distance 5 and angle 45
        
        glcm = graycomatrix(gray, [5], angles=[45], levels=256, symmetric=True, normed=True)
        
        contrast = graycoprops(glcm, prop="contrast")
        energy = graycoprops(glcm, prop='energy')
        homogeneity = graycoprops(glcm, prop='homogeneity')
        correlation = graycoprops(glcm, prop='correlation')
        
        contrast_5_45.append(contrast[0][0])
        energy_5_45.append(energy[0][0])
        homogeneity_5_45.append(homogeneity[0][0])
        correlation_5_45.append(correlation[0][0])
        
    data['path'] = path_df
    data['contrast_1_0'] = contrast_1_0
    data['energy_1_0'] = energy_1_0
    data['homogeneity_1_0'] = homogeneity_1_0
    data['correlation_1_0'] = correlation_1_0
    data['contrast_1_90'] = contrast_1_90
    data['energy_1_90'] = energy_1_90
    data['homogeneity_1_90'] = homogeneity_1_90
    data['correlation_1_90'] = correlation_1_90
    data['contrast_3_0'] = contrast_3_0
    data['energy_3_0'] = energy_3_0
    data['homogeneity_3_0'] = homogeneity_3_0
    data['correlation_3_0'] = correlation_3_0
    data['contrast_3_90'] = contrast_3_90
    data['energy_3_90'] = energy_3_90
    data['homogeneity_3_90'] = homogeneity_3_90
    data['correlation_3_90'] = correlation_3_90
    data['contrast_5_0'] = contrast_5_0
    data['energy_5_0'] = energy_5_0
    data['homogeneity_5_0'] = homogeneity_5_0
    data['correlation_5_0'] = correlation_5_0
    data['contrast_5_90'] = contrast_5_90
    data['energy_5_90'] = energy_5_90
    data['homogeneity_5_90'] = homogeneity_5_90
    data['correlation_5_90'] = correlation_5_90
    data['contrast_1_45'] = contrast_1_45
    data['energy_1_45'] = energy_1_45
    data['homogeneity_1_45'] = homogeneity_1_45
    data['correlation_1_45'] = correlation_1_45
    data['contrast_3_45'] = contrast_3_45
    data['energy_3_45'] = energy_3_45
    data['homogeneity_3_45'] = homogeneity_3_45
    data['correlation_3_45'] = correlation_3_45
    data['contrast_5_45'] = contrast_5_45
    data['energy_5_45'] = energy_5_45
    data['homogeneity_5_45'] = homogeneity_5_45
    data['correlation_5_45'] = correlation_5_45
    data['label'] = label
    
    return data
```

```python
df = pd.read_csv("new_dataset.csv")
df
```

```python
X = df.drop(["path","label"],axis=1)
X
```

```python
y = df["label"]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.33)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
```

```python
knn.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred_k = knn.predict(X_test_scaled)
```

```python
accuracy_score(y_test, y_pred_k)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train_scaled, y_train)
```

```python
y_pred_lr = lr_model.predict(X_test_scaled)
```

```python
accuracy_score(y_pred_lr, y_test)
```
